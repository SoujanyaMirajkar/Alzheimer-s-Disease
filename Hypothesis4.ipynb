```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from mpl_toolkits.mplot3d import Axes3D
data= pd.read_csv(r'/content/Alzheimer_Disease_notnull_final (1).csv')
data
```

# HYPOTHESIS

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data)

# Group data by ethnicity and calculate total prevalence for each group
ethnicity_total_df = df.groupby('Ethnicity').sum().reset_index()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(ethnicity_total_df['Data_Value'], labels=ethnicity_total_df['Ethnicity'], autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Alzheimer's Disease Prevalence by Ethnicity")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

### Is there a trend in Alzheimer's disease prevalence across different years (YearStart to YearEnd)?

```python
df = pd.DataFrame(data)

# Line chart showing Data_Value over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearEnd', y='Data_Value', data=df, marker='o')
plt.title("Alzheimer's Disease Prevalence Over Time")
plt.xlabel('Year')
plt.ylabel('Prevalence Rate')
plt.grid(True)
plt.show()
```

### Interaction between age and gender affects Alzheimer's disease prevalence.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.DataFrame(data)

# Interaction plot
plt.figure(figsize=(10, 6))
sns.pointplot(x='Age', y='Data_Value', hue='Gender', data=df, markers=['o', 's'])
plt.title("Interaction Between Age and Gender on Alzheimer's Disease Prevalence")
plt.xlabel('Age')
plt.ylabel('Prevalence Rate')
plt.show()
```

###  Distribution of Alzheimer's disease prevalence within each class

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.DataFrame(data)

# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Class', y='Data_Value', data=df)
plt.title("Distribution of Alzheimer's Disease Prevalence by Class")
plt.xlabel('Class')
plt.ylabel('Prevalence Rate')
plt.show()
```

What is the average (or median) prevalence of Alzheimer's disease across the entire dataset?


```python
# Calculate average and median prevalence
mean_prevalence = data['Data_Value'].mean()
median_prevalence = data['Data_Value'].median()

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(data['Data_Value'].dropna(), bins=30, alpha=0.7, color='blue')
plt.axvline(mean_prevalence, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_prevalence:.2f}')
plt.axvline(median_prevalence, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_prevalence:.2f}')
plt.title('Distribution of Alzheimer\'s Disease Prevalence')
plt.xlabel('Prevalence')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

How do the confidence limits vary across different ethni groups?

```python
# Group by 'Ethnicity' and calculate the mean of 'Low_Confidence_Limit' and 'High_Confidence_Limit'
ethnicity_grouped = data.groupby('Ethnicity')[['Low_Confidence_Limit', 'High_Confidence_Limit']].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))

# Bar plot for Low Confidence Limit
plt.bar(ethnicity_grouped['Ethnicity'], ethnicity_grouped['Low_Confidence_Limit'], color='blue', alpha=0.6, label='Low Confidence Limit')

# Bar plot for High Confidence Limit
plt.bar(ethnicity_grouped['Ethnicity'], ethnicity_grouped['High_Confidence_Limit'], color='red', alpha=0.6, label='High Confidence Limit', bottom=ethnicity_grouped['Low_Confidence_Limit'])

# Adding labels and title
plt.xlabel('Ethnicity')
plt.ylabel('Confidence Limit')
plt.title('Average Low and High Confidence Limits by Ethnicity')
plt.legend()

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()

# Print average values
print("Average Low and High Confidence Limits by Ethnicity:")
print(ethnicity_grouped)
```

What is the proportion of individuals with high and low confidence limits for prevalence estimates? (Data_Value_Type)

```python

import pandas as pd
import matplotlib.pyplot as plt



# Filter data for high and low confidence limits
high_confidence_data = data['High_Confidence_Limit']
low_confidence_data = data['Low_Confidence_Limit']

# Count occurrences
high_confidence_count = high_confidence_data.value_counts()
low_confidence_count = low_confidence_data.value_counts()

# Plotting for high confidence limit
plt.figure(figsize=(6, 6))
high_confidence_count.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['skyblue'])
plt.title('Proportion of High Confidence Limits')
plt.ylabel('')  # Hide y-label
plt.show()

# Plotting for low confidence limit
plt.figure(figsize=(6, 6))
low_confidence_count.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightgreen'])
plt.title('Proportion of Low Confidence Limits')
plt.ylabel('')  # Hide y-label
plt.show()
```
